```python
#usercase2: Display only those customers  from csv who purchases more than 3 products
```

```python
customers = pd.read_csv("customers.csv")
transactions = pd.read_csv("transactions.csv")
purchase_count = transactions.groupby("'custno'")["'productno'"].nunique().reset_index()
purchase_count.columns = ["'custno'",'num_products']
customers_more_than_3 = purchase_count[purchase_count['num_products'] > 3]
result = pd.merge(customers_more_than_3, customers, on="'custno'", how='left')
print(result)
```

```python
# usercase3: Dislay top 5 most demanding product from csv
```

```python
import pandas as pd

# Load CSV files
transactions = pd.read_csv('transactions.csv')
products = pd.read_csv('products.csv')

# Count frequency of each product sold
product_demand = transactions["'productno'"].value_counts().reset_index()
product_demand.columns = ["'productno'", "'frequency'"]

# Merge with products data to get product names
top_products = product_demand.merge(products, on="'productno'", how='left')

# Select top 5 most demanding products
top_5 = top_products.head(5)

# Display result
print("Top 5 Most Demanding Products:")
print(top_5[["'productno'", "'productName'", "'frequency'"]])
```
